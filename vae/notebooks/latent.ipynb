```python
import numpy as np
import matplotlib.pyplot as plt

# Path to the latent npy file
file_path = '/code/ggonsior/data/atd12k/test/Disney_0_00383_s2/latent1.npy'

# Load the latent file
latent = np.load(file_path)

print("Loaded latent file:", file_path)
```

```python
print("Shape:", latent.shape)
print("Data type:", latent.dtype)
print("Minimum value:", np.min(latent))
print("Maximum value:", np.max(latent))
print("Mean value:", np.mean(latent))
print("Standard deviation:", np.std(latent))
```

```python
plt.figure(figsize=(8, 4))
plt.hist(latent.flatten(), bins=50, color='skyblue', edgecolor='k')
plt.title("Histogram of Latent Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# If the latent file is 3-dimensional (e.g., [channels, height, width]),
# we can try to visualize it as an image.
if len(latent.shape) == 3:
    # Check if there are 3 channels; if not, we can just plot the first channel.
    if latent.shape[0] == 3:
        latent_img = latent.transpose(1, 2, 0)  # Rearrange to H x W x C
    else:
        latent_img = latent[0]  # Use the first channel
    plt.figure(figsize=(6, 6))
    plt.imshow(latent_img, cmap='gray')
    plt.title("Visual Inspection of Latent Representation")
    plt.axis("off")
    plt.show()
else:
    print("The latent file is not 3-dimensional, so it may not represent an image.")
```
